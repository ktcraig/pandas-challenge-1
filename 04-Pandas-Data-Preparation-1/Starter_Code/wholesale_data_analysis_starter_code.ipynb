## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
data_analysis_df['line_subtotal'] = data_analysis_df['unit_price'] * data_analysis_df['qty']
data_analysis_df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
"""SourceCode: ChatGPT assistance on writing lambda for weight calculation"""
data_analysis_df["total_weight"] = data_analysis_df["unit_weight"] * data_analysis_df["qty"]
data_analysis_df

data_analysis_df["shipping_price"] = data_analysis_df["total_weight"].apply(lambda weight: weight * 7 if weight > 50 else weight * 10)
data_analysis_df
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
sales_tax_rate = 0.0925

data_analysis_df["line_price"] = (data_analysis_df["line_subtotal"] + data_analysis_df["shipping_price"]) * (1 + sales_tax_rate)

data_analysis_df[["line_subtotal", "shipping_price", "line_price"]] 
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

data_analysis_df['line_cost'] = (data_analysis_df['unit_cost'] * data_analysis_df['qty']) + data_analysis_df["shipping_price"]
data_analysis_df[["unit_cost", "qty", "shipping_price", "line_cost"]] 
```

```python
# Create a column for the profit of each line using line cost and line price
data_analysis_df["line_profit"] = data_analysis_df["line_price"] - data_analysis_df["line_cost"]
data_analysis_df[["line_price", "line_cost", "line_profit"]]
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
final_data_analysis_df = data_analysis_df[['client_id', 'qty']].copy()
final_data_analysis_df.set_index('client_id', inplace=True)
final_data_analysis_df = final_data_analysis_df.sort_index()
result = final_data_analysis_df.loc[(33615, 66037, 46820, 38378, 24741), ["qty"]]

top_5_sum = result.groupby('client_id')['qty'].sum()
print(top_5_sum)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
"""Sourcecode: Xpert assistant helped to limit summary dataframe to only top 5 clients"""

summary_data_analysis_df = data_analysis_df[['client_id', 'qty', "shipping_price", "line_price", "line_profit"]].copy()
summary_data_analysis_df.set_index('client_id', inplace=True)


result = summary_data_analysis_df.loc[(33615, 66037, 46820, 38378, 24741), ["qty", "shipping_price", "line_price", "line_profit"]]


units_purchased = result.groupby('client_id')['qty'].sum()
total_shipping_price = result.groupby('client_id')['shipping_price'].sum()
total_revenue = result.groupby('client_id')['line_price'].sum()
total_profit = result.groupby('client_id')['line_profit'].sum()


summary_df = pd.DataFrame({
    'units_purchased': units_purchased,
    'total_shipping_price': total_shipping_price,
    'total_revenue': total_revenue,
    'total_profit': total_profit
})


summary_df = summary_df.loc[[33615, 66037, 46820, 38378, 24741]]


print(summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.
df_renamed = summary_df.rename(columns={"units_purchased": "Units Purchased",
                                                "total_shipping_price" : "Total Shipping",
                                                "total_revenue": "Total Revenue",
                                                "total_profit": "Total Profit"
                                                })
df_renamed.style.format("{:,.2f}", subset=['Units Purchased','Total Shipping','Total Revenue','Total Profit']) 


# Define the money columns. 
''''SourceCode: Tutoring Help from TA Luna 11/10'''

columns = ["Total Shipping", "Total Revenue", "Total Profit"]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(amount_of_dollars):
    """Function that converts dollars to millions."""
    amount_of_dollars = amount_of_dollars / 1000000
    return amount_of_dollars

# Apply the currency_format_millions function to only the money columns. 
for name in columns:
   
    df_renamed[name] = df_renamed[name].apply(currency_format_millions)
   
    
# Rename the columns to reflect the change in the money format. 
''''SourceCode: Tutoring Help from TA Luna 11/10'''
df_renamed = df_renamed.rename(columns={"Total Shipping" : "Total Shipping (in Millions) ",
                                                "Total Revenue" : "Total Revenue (in Millions)",
                                                "Total Profit" : "Total Profit (in Millions)"
                                                })

print(df_renamed)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_renamed_final = df_renamed.sort_values("Total Profit (in Millions)", ascending=False)

print(df_renamed_final)
```
